# Project Title: Athena AI Research Assistant

## Problem Description

The objective of this assignment was to design and implement an AI system with a conversational interface capable of answering user questions using Retrieval-Augmented Generation (RAG). Unlike a traditional large language model that relies solely on its pretrained knowledge, a RAG system retrieves relevant information from a document collection before generating a response. This approach improves factual accuracy, reduces hallucinations, and ensures that responses are grounded in the available knowledge base.

The application developed for this assignment is called **Athena AI Research Assistant**. It combines a Gradio-based conversational interface, a Chroma vector database, a Sentence Transformer embedding model, and the OpenAI GPT-4o-mini language model (accessed through the course API gateway). Users can ask questions in natural language, and the system retrieves the most relevant document(s) before generating an answer based on the retrieved context.

The chatbot currently answers questions about Artificial Intelligence and Retrieval-Augmented Generation using a small knowledge base stored in ChromaDB.


# System Architecture

The Athena AI Research Assistant follows a standard Retrieval-Augmented Generation (RAG) architecture.

**Workflow**

1. The user submits a question through the Gradio chat interface.
2. The question is converted into an embedding using the Sentence Transformer model (`all-MiniLM-L6-v2`).
3. ChromaDB performs a similarity search using the embedded query.
4. The most relevant document(s) are retrieved from the vector database.
5. A prompt is constructed by combining the retrieved context with the user's question.
6. The prompt is sent to the GPT-4o-mini language model through the course API gateway.
7. The generated response is returned to the Gradio interface.

# Implementation

The chatbot was implemented using the following components:

### User Interface

A conversational interface was developed using Gradio. The interface accepts natural language questions and displays the generated responses.

### Retrieval System

A Chroma persistent vector database was used to store document embeddings. The database performs semantic similarity search to retrieve the document most relevant to the user's query.

### Embedding Model

The `all-MiniLM-L6-v2` Sentence Transformer model was used to generate dense vector embeddings for both documents and user queries.

### Prompt Construction

The retrieved document is combined with the user's question using a prompt template defined in `prompts.py`. The prompt instructs the language model to answer only using the retrieved context.

### Language Model

Responses are generated using OpenAI's GPT-4o-mini model accessed through the course API gateway. The retrieved context is supplied as input to ensure that answers are grounded in the knowledge base.

### Knowledge Base

The current knowledge base consists of a sample document describing Artificial Intelligence, Retrieval-Augmented Generation, vector databases, embeddings, and the components of a RAG pipeline.

# Reflection

This assignment provided practical experience in designing and implementing a Retrieval-Augmented Generation application. Integrating document retrieval with a large language model demonstrated how external/provided knowledge can improve the factual accuracy of generated responses compared with relying solely on a pretrained language model.

Several implementation decisions were made during development. ChromaDB was selected as the vector database because it integrates well with Python and supports efficient semantic similarity search. The `all-MiniLM-L6-v2` Sentence Transformer was chosen because it provides high-quality embeddings while remaining lightweight enough for local execution. Gradio was selected for the conversational interface because it enables rapid development of interactive AI applications.

One challenge encountered during development was ensuring that the application consistently referenced the correct Chroma database. Additional work was required to resolve Python import paths, environment configuration, API gateway authentication, and vector database persistence. These issues were addressed by standardizing project paths, configuring the API gateway correctly, and rebuilding the vector database where necessary.

The current implementation is intentionally lightweight and serves as a proof of concept. Future improvements could include support for multiple documents, persistent conversation memory across sessions, more comprehensive guardrails for input validation and response safety, document upload capabilities, and evaluation metrics for measuring retrieval quality and response accuracy.

Overall, the assignment demonstrated the essential components of a modern RAG system, including document embedding, vector search, prompt construction, large language model integration, and conversational interaction through a user-friendly interface.
